```python
#YOLO_V3

YOLO_Area = []
for x in dict_YOLO:
    area = 0
    acc = []
    for obj in dict_YOLO[x]:
        area = area + (obj[3]-obj[1])*(obj[4]-obj[2])
        acc.append((obj[5]*100))
    YOLO_Area.append([area, np.mean(acc)])
```

```python
print(YOLO_Area)
```
